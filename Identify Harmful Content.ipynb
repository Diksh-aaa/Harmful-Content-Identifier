# **Data Preprocessing**


```python
from datasets import load_dataset, DatasetDict, ClassLabel

# Custom label assignment function
def assign_custom_label(example, dataset_name):
    if dataset_name == "DarkyMan/nsfw-image-classification":
        example["label"] = 1  # Custom label for DarkyMan/nsfw-image-classification
    elif dataset_name == "x1101/nsfw-full":
        example["label"] = 2  # Custom label for x1101/nsfw-full
    elif dataset_name == "NeuralShell/Gore-Blood-Dataset-v1.0":
        example["label"] = 3  # Custom label for NeuralShell/Gore-Blood-Dataset-v1.0
    elif dataset_name == "rtb1271/human_fight_detection":
        example["label"] = 4  # Custom label for rtb1271/human_fight_detection
    else:
        example["label"] = 0  # Default label
    return example

# Function to add label column and redefine ClassLabel feature
def add_label_column_and_redefine_class_label(dataset, num_classes):
    # Remove existing 'labels' column if it exists
    if 'labels' in dataset['train'].features:
        dataset = dataset.remove_columns('labels')

    # Add a label column initialized to 0
    dataset = dataset.map(lambda example: {"label": 0})

    # Create a new ClassLabel feature with the appropriate number of classes
    new_features = dataset["train"].features.copy()
    new_features["label"] = ClassLabel(num_classes=num_classes)

    # Apply the new feature to all splits in the dataset
    dataset = dataset.cast(new_features)
    return dataset

# Function to load and process a dataset
def load_and_process_dataset(dataset_name, num_classes):
    dataset = load_dataset(dataset_name)

    # Add label column and redefine the class label feature
    dataset = add_label_column_and_redefine_class_label(dataset, num_classes)

    # Apply custom label assignment
    dataset = dataset.map(lambda example: assign_custom_label(example, dataset_name))

    # Split the dataset into training and validation sets
    train_test_split = dataset['train'].train_test_split(test_size=0.2, seed=42)
    split_dataset = DatasetDict({
        'train': train_test_split['train'],
        'validation': train_test_split['test']
    })
    return split_dataset

# Number of classes needed (including the default label 0)
num_classes = 5

# Load and process all datasets
datasets = ["DarkyMan/nsfw-image-classification", "x1101/nsfw-full", "NeuralShell/Gore-Blood-Dataset-v1.0", "rtb1271/human_fight_detection"]
processed_datasets = {name: load_and_process_dataset(name, num_classes) for name in datasets}

# Print the processed datasets
for name, dataset in processed_datasets.items():
    print(f"Dataset: {name}")
    print(dataset)
    print(dataset["train"].features)

# Example output to verify labels
for name, dataset in processed_datasets.items():
    print(f"Sample from {name}:")
    print(dataset['train'][0])  # Print the first example from the training set
```

# **Training Model**

The line ImageFile **'LOAD_TRUNCATED_IMAGES = True'** from the **'PIL'** library (Python Imaging Library) is used to handle truncated images.



```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from datasets import concatenate_datasets
from PIL import Image, ImageFile

# Allow truncated images to be loaded
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Combine datasets
def combine_datasets(datasets):
    combined_train = concatenate_datasets([datasets[name]['train'] for name in datasets])
    combined_validation = concatenate_datasets([datasets[name]['validation'] for name in datasets])
    return combined_train, combined_validation

# Custom Dataset class for PyTorch
class CustomDataset(Dataset):
    def __init__(self, hf_dataset, transform=None):
        self.dataset = hf_dataset
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]
        image = sample['image']
        label = sample['label']

        # Convert image to RGB
        if image.mode != 'RGB':
            image = image.convert('RGB')

        if self.transform:
            image = self.transform(image)
        return image, label

# Transformations for the images
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Combine all datasets
combined_train, combined_validation = combine_datasets(processed_datasets)

# Create PyTorch datasets
train_dataset = CustomDataset(combined_train, transform=transform)
validation_dataset = CustomDataset(combined_validation, transform=transform)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
validation_loader = DataLoader(validation_dataset, batch_size=16, shuffle=False)

# Define a simple CNN model
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=5):
        super(SimpleCNN, self).__init__()
        self.model = models.resnet18(weights='ResNet18_Weights.IMAGENET1K_V1')
        self.model.fc = nn.Linear(self.model.fc.in_features, num_classes)

    def forward(self, x):
        return self.model(x)

# Initialize the model, loss function, and optimizer
model = SimpleCNN(num_classes=5)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training function
def train_model(model, train_loader, validation_loader, criterion, optimizer, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        running_corrects = 0
        total_train = 0
        for images, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * images.size(0)
            _, preds = torch.max(outputs, 1)
            running_corrects += torch.sum(preds == labels.data)
            total_train += labels.size(0)

        epoch_loss = running_loss / total_train
        epoch_acc = running_corrects.double() / total_train

        model.eval()
        val_running_loss = 0.0
        val_running_corrects = 0
        total_val = 0
        with torch.no_grad():
            for images, labels in validation_loader:
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_running_loss += loss.item() * images.size(0)
                _, preds = torch.max(outputs, 1)
                val_running_corrects += torch.sum(preds == labels.data)
                total_val += labels.size(0)

        val_epoch_loss = val_running_loss / total_val
        val_epoch_acc = val_running_corrects.double() / total_val

        print(f'Epoch {epoch+1}/{num_epochs}')
        print(f'Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
        print(f'Val Loss: {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}')
        print('-' * 20)

# Train the model
train_model(model, train_loader, validation_loader, criterion, optimizer, num_epochs=3)
```
